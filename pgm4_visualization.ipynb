```python
import matplotlib.pyplot as plt
```

```python
x = [1, 4, 3]
y = [2, 4, 1]
plt.plot(x, y, linewidth = 4)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('My first graph')
plt.show()
```

```python
x = [ 1, 2, 3]
y = [2, 4, 1]
plt.bar(x, y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title("My first graph")
plt.show()
```

```python
y = [10, 5, 8, 4, 2, 2, 2, 5]
plt.hist(y)
plt.show()
```

```python
x = [5, 2, 9, 4, 7]
y = [10, 5, 8, 4, 2]
plt.scatter(x, y)
plt.show()
```

# Data Visualization using ploty

```python
import plotly.express as px
fig = px.line(x=[1,2,3], y=[2,4,6])
fig.show()
```

```python
df = px.data.iris()
print(df)
fig = px.line(df, x="species", y="petal_width")
fig.show()
```

```python
fig = px.bar(df, x="sepal_width", y="sepal_length")
fig.show()
```

```python
fig = px.histogram(df, x = "sepal_length", y = "petal_width")
fig.show()
```

```python
fig = px.scatter(df, x="species", y="petal_width")
fig.show()
```

```python
fig = px.scatter(df, x="species", y="petal_width", size = "petal_length", color = "species")
fig.show
```

```python
fig = px.scatter_3d(df, x= "sepal_width", y = "sepal_length", z = "petal_width", color = "species")
fig.show()
```
